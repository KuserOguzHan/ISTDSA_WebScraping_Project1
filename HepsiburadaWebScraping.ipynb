#1. WEB SCRAPING

```python
from bs4 import BeautifulSoup as bts
import requests
import pandas as pd
import numpy as np
```

```python
def getAndParseURL(url):
  result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
  soup = bts(result.text,"html.parser")
  return soup
```

```python
getAndParseURL("https://www.hepsiburada.com/cep-telefonlari-c-371965")
```

```python
PAGE_NUM = []
def find_page_num():
    

    html = getAndParseURL("https://www.hepsiburada.com/cep-telefonlari-c-371965")
    
    for i in range(0,50):
 
        PAGE_NUM.append("https://www.hepsiburada.com/cep-telefonlari-c-371965"+str("?sayfa=")+str(i))
        print(PAGE_NUM)

    return PAGE_NUM
find_page_num()
```

```python
ALL_PRODUCT = []
def all_product():
    
    for i in PAGE_NUM[::]:
      html = getAndParseURL(i)
      for links in html.find_all("li",{"class": "productListContent-item"}):
        ALL_PRODUCT.append("https://www.hepsiburada.com/"+links.a["href"])
        print("https://www.hepsiburada.com/"+links.a["href"])

    return ALL_PRODUCT

all_product()
```

```python
len(ALL_PRODUCT)
```

```python
result = []
for details in ALL_PRODUCT[::]:
    html = getAndParseURL(details)
    
    #name

    nm = html.find("h1", {"itemprop":"name"})
                              
    if nm is None:
      name= np.NaN
    
    else:
      name = nm.text.strip()

    #ram
    rm = html.find("div", class_="key-properties").find(text=" RAM  Kapasitesi ")
                              
    if rm is None:
      ram= np.NaN
    
    else:
      ram = rm.findNext().text.strip()


    #memory  
    mm = html.find("div", class_="key-properties").find(text=" Dahili Hafıza ")
                              
    if mm is None:
      memory= np.NaN
    
    else:
      memory = mm.findNext().text.strip()   
  

    #scren
    sc =html.find("div", class_="key-properties").find(text=" Ekran Boyutu ")

    if sc is None:
      screen= np.NaN
    
    else:
      screen= sc.findNext().text.strip()    


    #power
    pw = html.find("div", class_="key-properties").find(text=" Pil Gücü ")
     
    if pw is None:
      power= np.NaN
    
    else:
      power= pw.findNext().text.strip()    
   
    
    
    #front_camera
    fc = html.find("div", class_="key-properties").find(text=" Ön (Selfie) Kamera ")
   
    if fc is None:
      front= np.NaN
    
    else:
      front= fc.findNext().text.strip()



    #rear_camere
    rc = html.find("div", class_="key-properties").find(text=" Kamera Çözünürlüğü ")
  
    if rc is None:
      rear= np.NaN
    
    else:
      rear= rc.findNext().text.strip()

    #price
    price = 0
    for i in html.findAll("div",{"class":"price-wrapper visible"}) :
            for b in i.findAll("del",{"class":"price-old"}) :
             price = (b.text)


    result.append([name, ram, memory, screen, power, front, rear, price])



columns = ["name","ram","memory","screen_size","power","front_camera","rear_camera", "price"]
df = pd.DataFrame.from_records(result, columns=columns)
```
